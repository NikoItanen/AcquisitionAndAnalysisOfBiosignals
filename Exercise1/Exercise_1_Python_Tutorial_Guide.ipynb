## Indexing and Slicing
* Note that we can access elements of a string, tuple, or list by integer indexing. Python indexing starts at 0.
* Access a slice of a list or tuple by specifying a slice of said list!

```python
colors = ['red', 'green', 'blue', 'yellow', 'white', 'black']

print (colors[0],"\n") # select one element from the list

print (colors[0:3],"\n") # select a slice from the list

print (colors[5][0:2],"\n")# select one element and a slice of that element because it's a string

print (colors[-1],colors[-6]) #negative indexing
```

## SciPy
* SciPy is a collection of mathematical algorithms and convenience functions built on the NumPy extension of Python. 
* Has a variety of different modules that focus on topics such as signal processing, linear algebra, statistics, and more. 

```python
Fs=200
sig_len=len(ECG_sig)
ECG_meanremoved=ECG_sig-np.mean(ECG_sig)


fx=fftpack.rfftfreq(sig_len,1/Fs) #get freqs for x axis
fy=fftpack.rfft(ECG_sig) # get real value fft, from 0hz to Fs/2
plt.figure(figsize=(12,7))
plt.title("Frequency Domain of our Example ECG Signal")
plt.xlabel("Frequency",fontsize=16)
plt.ylabel("Amplitude",fontsize=16)
plt.plot(fx,np.abs(fy))

```

* What you see is the frequency content of the ECG signal we have.
* In order to better visualize the frequency content of the signal, remove the DC frequency (frequency = 0) and replot the frequency content.
* Show the new plot to the instructor. How many peaks do you see there? what are these peaks representing?  

 
* HINT: DC frequency can be removed by subtracting the mean from your signal

```python
#answer using scipy
Fs=200
Nyquist_freq=Fs/2 # max frequency that can be present in our signal
sig_len=len(ECG_sig)
ECG_meanremoved=ECG_sig-np.mean(ECG_sig)

fx=fftpack.rfftfreq(sig_len,1/Fs)
fy=np.abs(fftpack.rfft(ECG_meanremoved))

# find index of top 3 peaks
ind= np.argpartition(np.abs(fy), -3)[-3:]

plt.figure(figsize=(12,7))
plt.title("Frequency Domain of our Example ECG Signal with DC frequency Removed")
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.plot(fx,np.abs(fy))

#plot red x's where the top 3 peaks are
plt.plot(fx[ind],fy[ind],"rx")
plt.show()
```

* Can we identify what frequency peak could be associated with heart rate?

```python
#view of low frequency components
ind= np.argpartition(np.abs(fy), -3)[-3:]
plt.plot(fx[:int(len(fx)/15)],np.abs(fy)[:int(len(fx)/15)])
plt.plot(fx[ind],np.abs(fy[ind]),"rx")

print (sorted(fx[ind]))
```
